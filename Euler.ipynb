```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
# Define ODE
#f = lambda t, s: np.exp(-t) 
f = lambda t, s: 4/(1+s**2)
#f = lambda t, s: t + s/5
```

```python
# Define parameters
h = 0.5 # Step size
t = np.arange(0, 4+h, h) # Numerical grid
s0 = 0 # Initial Condition
```

```python
# Numericall integrate ODE using off-the-shelf commands
sol = solve_ivp(f, t, [s0], method='DOP853', dense_output=True)
z = sol.sol(t)
```

```python
#Euler method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])
```

```python
# Plotting the results.
plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, z.T, 'g', label='Off the shelf command')
plt.title('Eulers method')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
